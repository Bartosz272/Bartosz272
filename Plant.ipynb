```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import asarray
```

```python
plant = plt.imread('Plant.jpg')
print(plant.shape)
plt.imshow(plant)
```

```python
small_plant = plt.imread('Small_Plant.jpg')
print(small_plant.shape)
plt.imshow(small_plant)
```

```python
forest = plt.imread('Forest.jpg')
print(forest.shape)
plt.imshow(forest)
```

```python
green = plt.imread('green.jpg')
print(green.shape)
plt.imshow(green)
```

```python
red = plt.imread('red.jpg')
print(red.shape)
plt.imshow(red)
```

```python
plant_array = asarray(plant)
small_plant_array = asarray(small_plant)
forest_array = asarray(forest)
green_array = asarray(green)
red_array = asarray(red)

number_of_green_pix_plant = np.sum(plant_array[:,:,1], where=((plant_array[:,:,0]*2 < plant_array[:,:,1]) & (plant_array[:,:,2]*2 < plant_array[:,:,1])))
plant_pixel = plant_array.shape[0]*plant_array.shape[1] *255
print(f'{(number_of_green_pix_plant/plant_pixel)*100}% of green in the image of a plant')

number_of_green_pix_small_plant = np.sum(small_plant_array[:,:,1], where=((small_plant_array[:,:,0]*2 < small_plant_array[:,:,1]) & (small_plant_array[:,:,2]*2 < small_plant_array[:,:,1])))
small_plant_pixel = small_plant_array.shape[0]*small_plant_array.shape[1] *255
print(f'{(number_of_green_pix_small_plant/small_plant_pixel)*100}% of green in the image of a small plant')

number_of_green_pix_forest = np.sum(forest_array[:,:,1], where=((forest_array[:,:,0]*2 < forest_array[:,:,1]) & (forest_array[:,:,2]*2 < forest_array[:,:,1])))
forest_pixel = forest_array.shape[0]*forest_array.shape[1] *255
print(f'{(number_of_green_pix_forest/forest_pixel)*100}% of green in the image of a forest')

number_of_green_pix_green = np.sum(green_array[:,:,1], where=((green_array[:,:,0]*2 < green_array[:,:,1]) & (green_array[:,:,2]*2 < green_array[:,:,1])))
green_pixel = green_array.shape[0]*green_array.shape[1] *255
print(f'{(number_of_green_pix_green/green_pixel)*100}% of green in the image of green')

number_of_green_pix_red = np.sum(red_array[:,:,1], where=((red_array[:,:,0]*2 < red_array[:,:,1]) & (red_array[:,:,2] < red_array[:,:,1]*2)))
red_pixel = red_array.shape[0]*red_array.shape[1] *255
print(f'{(number_of_green_pix_red/red_pixel)*100}% of green in the image of red')


```
